```python
mapping_table={5:7.0,10:6.7,15:6.3,20:6.0,30:5.7,40:5.3,50:5.0,60:4.8,70:4.6,80:4.4,90:4.2,100:4,120:4.5,130:4.2}
def calc_feed(count,total_weight):
  avg_weight=total_weight/(count)
  print(avg_weight)
  aw = round(avg_weight/10)*10
  print(mapping_table[aw])
  
  return count*avg_weight*mapping_table[aw]/100
```

```python
def Rand(start, end, num):
    res = []
 
    for j in range(num):
        res.append(random.uniform(start, end))
 
    return res
```

```python
from google.colab.patches import cv2_imshow

@app.route('/')
def home():
    return render_template('index.html')
@app.route('/getprediction',methods=['POST'])

def getprediction():

  if request.method == 'POST':  

        
        f = request.files['file']
        f.save("./image.png")
        image = cv2.imread("image.png")
        response=YOLO('best1.pt')(image)
        print(response)
        r=response[0].plot(boxes=True)
        print(response[0].boxes.data) 
        print("*******")
        print(response[0].boxes[0].xyxy[0][0])
        print(response[0].boxes[0].xyxy)
        
        cv2_imshow(r)
        cv2.imwrite('./static/yoloboundingbox.png', r)
        
        depth_estimator = joblib.load('depth.pkl')
        p_cm_convertor=joblib.load('p_cm.pkl')
        biomass_estimator=joblib.load('biomass.pkl')
        
        length_pix=[]
        width_pix=[]
        for i in range(len(response[0].boxes)):
          width_pix.append(abs(response[0].boxes[i].xyxy[0][2]-response[0].boxes[i].xyxy[0][0]))
          length_pix.append(abs(response[0].boxes[i].xyxy[0][3]-response[0].boxes[i].xyxy[0][1]))
          
          

        df = pandas.DataFrame({'length_pix': length_pix,'width_pix':width_pix})
                # Use the loaded model to make predictions
        depth_result=depth_estimator.predict(df)


        df=pandas.DataFrame({'length_pix': length_pix,'width_pix':width_pix, 'depth':depth_result})
        p_cm_result=p_cm_convertor.predict(df)


        df1 = pandas.DataFrame(p_cm_result, columns = ['length_cm','width_cm'])
        turbidity=Rand(10,15,len(p_cm_result))
        df2=pandas.DataFrame(turbidity,columns =['turbidity'])
        print(type(turbidity))
        horizontal_concat = pandas.concat([df1, df2], axis=1)
        biomass=biomass_estimator.predict(horizontal_concat)


        #df=pandas.DataFrame({'length_pix': [5,6],'width_pix':[10,8], 'depth':depth_result})
        #print(p_cm_result[0])

        print(biomass)
        count=len(biomass)
        tot_weight=0
        for i in range(count):
          tot_weight+=biomass[i]
        
        print(tot_weight)
        print(round(calc_feed(count,tot_weight),3))
        feed=round(calc_feed(count,tot_weight),3)
        tot_weight=round(tot_weight,3)
        print("*****")
        
        

  return render_template('result.html', user_image='./static/yoloboundingbox.png',biomass=tot_weight,feed=feed)

app.run()
```
